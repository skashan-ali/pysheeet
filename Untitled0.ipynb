```python
from sklearn.datasets import fetch_20newsgroups
from sentence_transformers import SentenceTransformer
from bertopic import BERTopic
from umap import UMAP
import pandas as pd
```

```python
# Load data
docs = pd.read_csv('./tmdb_150_movies.csv')[:148]
overview = docs['overview']
```

```python
docs
```

```python
sentence_model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
embeddings = sentence_model.encode(overview, show_progress_bar=False)
```

```python
embeddings
```

```python
# Train BERTopic
topic_model = BERTopic().fit(overview, embeddings)

# Run the visualization with the original embeddings
topic_model.visualize_documents(overview, embeddings=embeddings)

# Reduce dimensionality of embeddings, this step is optional but much faster to perform iteratively:
reduced_embeddings = UMAP(n_neighbors=10, n_components=2, min_dist=0.0, metric='cosine').fit_transform(embeddings)
```

```python
topic_model.visualize_documents(docs['original_title'], reduced_embeddings=reduced_embeddings)
```
